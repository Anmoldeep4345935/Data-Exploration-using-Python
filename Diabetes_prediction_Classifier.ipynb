```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

```python
df = pd.read_csv("Dia.csv")
df
```

```python
df = df.drop([], axis=1)
df.head(5)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna() # Dropping the missing values.
df.count()
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
print("\nSummary statistics:")
print(df.describe())
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(x='class', data=df)
plt.title('Distribution of Classes')
plt.show()
```

```python
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_features].hist(figsize=(12, 10))
plt.suptitle('Histograms of Numeric Features', x=0.5, y=1.02, ha='center', fontsize='large')
plt.show()
```

```python


X = df.drop('class', axis=1)
y = df['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='bar')
plt.title('Feature Importances')
plt.show()
```

```python

sns.scatterplot(x='Tri Fold Thick', y='BMI', data= df)
plt.title('Scatter Plot of Tri Fold Thick vs BMI')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python

sns.histplot(data=df, x='Tri Fold Thick', bins=20, color='blue', alpha=0.9, label='Tri Fold Thick')
sns.histplot(data=df, x='BMI', bins=22, color='green', alpha=0.9, label='BMI')
plt.title('Histogram of Tri Fold Thick and BMI')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
categories = list(df)[1:]
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()

# Close the plot
df = df.append(df.iloc[0])

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories, color='black', size=10)

# Draw ylabels
ax.set_rscale('log')
plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.5", "0.75"], color="grey", size=7)
plt.ylim(0, 1)

# Plot data
for i in range(len(df)):
    values = df.loc[i].drop('Category').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=df['Category'][i])

# Fill area
ax.fill(angles, values, 'b', alpha=0.1)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
